# MP4 视频转 GIF 转换器

这个Jupyter笔记本将帮助你将MP4视频文件转换为GIF动画格式。GIF格式适合在网站和社交媒体上分享动画内容。

## 第1部分: 导入所需库

导入必要的库，包括OpenCV(cv2)、PIL(Pillow)和NumPy，用于视频处理和图像转换。

```python
import cv2
import numpy as np
from PIL import Image
import os
from pathlib import Path

print("✓ 所有库导入成功！")
```

## 第2部分: 加载视频文件

使用OpenCV读取视频文件，获取视频的基本信息，如帧数、分辨率和帧率。

```python
# 设置视频路径
video_dir = r"d:\MASTER\GITPRO\MGDF\images"
video_files = ["1.mp4", "2.mp4", "3.mp4", "4.mp4", "5.mp4", "6.mp4", "7.mp4", "8.mp4"]

# 加载第一个视频文件进行演示
video_path = os.path.join(video_dir, video_files[0])

# 打开视频文件
cap = cv2.VideoCapture(video_path)

# 获取视频信息
fps = cap.get(cv2.CAP_PROP_FPS)  # 帧率
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))  # 总帧数
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  # 宽度
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # 高度
duration = frame_count / fps  # 总时长（秒）

print(f"视频文件: {os.path.basename(video_path)}")
print(f"分辨率: {width}x{height}px")
print(f"帧率: {fps} fps")
print(f"总帧数: {frame_count}")
print(f"总时长: {duration:.2f}秒")
```

## 第3部分: 提取视频帧

逐帧读取视频，提取所需的帧数据，可以选择每N帧提取一次以减少GIF大小。

```python
# 参数配置
frame_skip = max(1, int(fps / 10))  # 每秒10帧（降采样）
frames = []
frame_index = 0

print(f"开始提取帧数据...（每{frame_skip}帧提取一次）")

cap = cv2.VideoCapture(video_path)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # 每N帧提取一次
    if frame_index % frame_skip == 0:
        frames.append(frame)
    
    frame_index += 1

cap.release()

print(f"✓ 提取完成！总共提取 {len(frames)} 帧")
print(f"原始帧数: {frame_count}, 提取帧数: {len(frames)}")
```

## 第4部分: 调整帧大小和优化

将提取的帧调整为合适的分辨率，转换颜色空间从BGR到RGB，优化图像质量和文件大小。

```python
# 调整帧大小
target_width = 400  # GIF宽度
scale_ratio = target_width / width
target_height = int(height * scale_ratio)

processed_frames = []

print(f"调整帧大小: {width}x{height} -> {target_width}x{target_height}")

for i, frame in enumerate(frames):
    # 转换颜色空间：BGR -> RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # 调整大小
    frame_resized = cv2.resize(frame_rgb, (target_width, target_height), 
                               interpolation=cv2.INTER_LANCZOS4)
    
    # 转换为PIL Image
    pil_image = Image.fromarray(frame_resized)
    processed_frames.append(pil_image)
    
    if (i + 1) % 10 == 0:
        print(f"已处理 {i + 1}/{len(frames)} 帧")

print(f"✓ 帧处理完成！")
```

## 第5部分: 创建GIF动画

使用PIL将处理后的帧序列转换为GIF格式，设置合适的帧延迟时间(duration)和循环参数。

```python
# 计算每帧的显示时间（毫秒）
frame_duration = int(1000 / (fps / frame_skip))

print(f"创建 GIF 动画...")
print(f"帧延迟: {frame_duration}ms")

# 创建GIF（保存第一帧，然后追加其他帧）
if processed_frames:
    gif = processed_frames[0]
    gif_params = {
        'save_all': True,
        'append_images': processed_frames[1:],
        'duration': frame_duration,  # 每帧显示时间（毫秒）
        'loop': 0,  # 0 = 无限循环
        'optimize': False,  # 不优化以保持质量
    }
    
    print(f"✓ GIF 动画已准备（{len(processed_frames)} 帧）")
```

## 第6部分: 保存GIF文件

将生成的GIF动画保存到指定路径，并验证输出文件。

```python
# 保存GIF文件
output_filename = "1.gif"
output_path = os.path.join(video_dir, output_filename)

print(f"保存GIF到: {output_path}")

gif.save(
    output_path,
    **gif_params
)

# 验证文件
if os.path.exists(output_path):
    file_size_mb = os.path.getsize(output_path) / (1024 * 1024)
    print(f"✓ GIF 保存成功！")
    print(f"文件大小: {file_size_mb:.2f} MB")
    print(f"输出路径: {output_path}")
else:
    print(f"✗ 保存失败！")
```

## 第7部分: 批量转换所有视频

将上述过程应用于所有视频文件，实现批量转换。

```python
def convert_video_to_gif(video_path, output_path, target_width=300, target_fps=5):
    """
    将视频文件转换为GIF动画（优化版本）
    
    参数:
        video_path: 输入视频文件路径
        output_path: 输出GIF文件路径
        target_width: GIF宽度（像素）- 默认300（更小）
        target_fps: 目标帧率（每秒帧数）- 默认5（更低）
    
    返回:
        success: 是否转换成功
        message: 转换消息
    """
    try:
        # 打开视频
        cap = cv2.VideoCapture(video_path)
        fps = cap.get(cv2.CAP_PROP_FPS)
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        # 提取帧
        frame_skip = max(1, int(fps / target_fps))
        frames = []
        frame_index = 0
        
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            if frame_index % frame_skip == 0:
                # 调整大小并转换颜色
                scale_ratio = target_width / width
                target_height = int(height * scale_ratio)
                
                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                # 转换为8-bit RGB图像，只保留256种颜色
                img = Image.fromarray(frame_rgb)
                img = img.quantize(colors=256)  # 颜色量化
                
                # 调整大小
                img = img.resize((target_width, target_height), 
                                 Image.Resampling.LANCZOS)
                frames.append(img)
            
            frame_index += 1
        
        cap.release()
        
        # 保存为GIF（启用优化）
        if frames:
            frame_duration = int(1000 / (fps / frame_skip))
            frames[0].save(
                output_path,
                save_all=True,
                append_images=frames[1:],
                duration=frame_duration,
                loop=0,
                optimize=True,  # 启用优化以减少文件大小
                quality=85  # 降低质量以减少文件大小
            )
            
            file_size_mb = os.path.getsize(output_path) / (1024 * 1024)
            return True, f"✓ {os.path.basename(output_path)} ({file_size_mb:.2f}MB) - {len(frames)}帧"
        
        return False, "No frames extracted"
    
    except Exception as e:
        return False, f"✗ 错误: {str(e)}"

# 批量转换所有视频
print("开始批量转换所有视频文件...\n")
print("优化参数: 宽度300px, 帧率5fps, 优化启用\n")

results = []
for video_file in video_files:
    video_path = os.path.join(video_dir, video_file)
    gif_file = video_file.replace(".mp4", ".gif")
    output_path = os.path.join(video_dir, gif_file)
    
    if os.path.exists(video_path):
        success, message = convert_video_to_gif(video_path, output_path)
        results.append((video_file, success, message))
        print(message)
    else:
        print(f"✗ 文件不存在: {video_file}")
```
